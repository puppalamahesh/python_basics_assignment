```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go

# Set random seed for reproducibility
np.random.seed(20)

# Generate sample sales data
data1 = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

# Create DataFrame
df1 = pd.DataFrame(data1)

# Create a pivot table to organize data by Day and Month
pivot_df = df1.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean').fillna(0)

# Create heatmap
fig1 = go.Figure(data=[go.Heatmap(
    z=pivot_df.values,    # Heatmap values from pivot table
    x=pivot_df.columns,   # Months as x-axis
    y=pivot_df.index,     # Days as y-axis
    colorscale='Viridis'
)])

# Set axis titles for clarity
fig1.update_layout(
    title="Sales Heatmap by Month and Day",
    xaxis_title="Month",
    yaxis_title="Day"
)

# Display the plot
fig1.show()
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go

# Set random seed for reproducibility
np.random.seed(25)

# Create sample data
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}

# Create DataFrame
df = pd.DataFrame(data)

# Create bubble chart
fig = go.Figure(data=[go.Scatter(
    x=df['GDP'],               # GDP on x-axis
    y=df['Population'],        # Population on y-axis
    text=df['Country'],        # Country names as hover text
    mode='markers',            # Bubble chart mode
    marker=dict(
        size=df['Population'], # Bubble size proportional to Population
        sizemode='area',       # Adjusts bubble size by area
        sizeref=2.*max(df['Population'])/(100**2), # Controls the scaling of bubbles
        color=df['Population'], # Color can also represent population
        colorscale='Viridis',  # Color scheme
        showscale=True         # Show color scale
    )
)])

# Add chart title and axis labels
fig.update_layout(
    title="Bubble Chart of Country Population vs. GDP",
    xaxis_title="GDP (in Billion USD)",
    yaxis_title="Population (in Millions)"
)

# Show the figure
fig.show()
```

**1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x?**

```python
from bokeh.plotting import figure, show
import numpy as np
from bokeh.io import output_notebook

# Enable Bokeh output in a Jupyter Notebook
output_notebook()

# Generate data
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.sin(x)                # Sine values for each x

# Create a new plot with title and axis labels
p = figure(title="Sine Wave", x_axis_label='X', y_axis_label='sin(X)')

# Add a line renderer with a legend and line thickness
p.line(x, y, legend_label="y = sin(x)", line_width=2, color="blue")

# Display the plot
show(p)
```

**2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the

**3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.**

 fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']

 counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource

# Enable Bokeh output in a Jupyter Notebook
output_notebook()

# Define the data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a data source
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a new figure
p = figure(
    x_range=fruits,                  # Use fruits as x-axis categories
    title="Fruit Counts",
    x_axis_label="Fruit",
    y_axis_label="Count",
    toolbar_location=None,
    tools=""
)

# Add a bar renderer with custom colors and data source
p.vbar(
    x='fruits',
    top='counts',
    width=0.5,
    source=source,
    legend_field="fruits",
    line_color='white',
    fill_color=factor_cmap('fruits', palette=['#ff9999','#66b3ff','#99ff99','#ffcc99'], factors=fruits)
)

# Customize the legend and layout
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show the plot
show(p)
```

**4. Create a Bokeh histogram to visualize the distribution of the given data.
 data_hist = np.random.randn(1000)
 hist, edges = np.histogram(data_hist, bins=30)**

```python
 import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Enable Bokeh output in a Jupyter Notebook
output_notebook()

# Generate data and histogram bins
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a data source for Bokeh
source = ColumnDataSource(data=dict(
    top=hist,
    left=edges[:-1],     # Left edges of the bins
    right=edges[1:]      # Right edges of the bins
))

# Create the figure for the histogram
p = figure(title="Histogram of Data Distribution",
           x_axis_label="Value",
           y_axis_label="Frequency",
           background_fill_color="#fafafa")

# Plot the histogram using the quad glyph
p.quad(
    top='top',
    bottom=0,
    left='left',
    right='right',
    fill_color="blue",
    line_color="white",
    source=source
)

# Show the plot
show(p)
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.transform import linear_cmap

# Enable Bokeh output in a Jupyter Notebook
output_notebook()

# Generate the data for the heatmap
data_heatmap = np.random.rand(10, 10)  # 10x10 grid of random values

# Set up a color mapper for the heatmap
color_mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max())

# Create the figure for the heatmap
p = figure(
    title="Bokeh Heatmap",
    x_axis_label="X Axis",
    y_axis_label="Y Axis",
    tooltips=[("Value", "@image")],
    width=400,
    height=400
)

# Plot the heatmap using the image glyph
p.image(
    image=[data_heatmap],        # Use the data_heatmap as the image source
    x=0,                         # Specify the x-position
    y=0,                         # Specify the y-position
    dw=1,                        # Width of the plot in data space
    dh=1,                        # Height of the plot in data space
    color_mapper=color_mapper    # Map colors to data values
)

# Add a color bar for the heatmap
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
